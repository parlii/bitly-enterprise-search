````python
from vertexai.language_models import TextGenerationModel
import requests
from bs4 import BeautifulSoup
````

````python
model = TextGenerationModel.from_pretrained('text-bison@latest')
````

## Generate custom back half

````python
response = requests.get('https://www.thepioneerwoman.com/food-cooking/meals-menus/g33025738/fall-recipes/')
````

````python
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text()
````

````python
prompt = f"""Given the following text extracted from a website, delimited by triple backticks:

```{text}```

Come up with a short, memorable, creative, less than 100 character, url friendly link name that can be used in a URL shortener.

Answer: """
````

````python
result = model.predict(prompt)
````

````python
print(result.text)
````
